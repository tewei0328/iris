```
features=iris['data']
```

```
labels=iris['target']
```

```
from sklearn import tree 
#features = [[150,1],[170,1],[130,0],[140,0]]
#labels = [0,0,1,1]

clf = tree.DecisionTreeClassifier()
clf = clf.fit(features,labels)

```

```
wantPredict = clf.predict([[5.1, 3.5, 1.4, 0.2]]) 
wantPredict
```

```
wantPredict = clf.predict([features[50]]) 
wantPredict
```

```
wantPredict = clf.predict([features[100]]) 
wantPredict
```

```
import pandas as pd
x = pd.DataFrame(iris['data'], columns=iris['feature_names'])
x
```

```

y = pd.DataFrame(iris['target'], columns=['target_names'])
y
```

```
df = pd.concat([x,y], axis=1)
df
```

```
import seaborn as sb
#iris=sb.load_dataset('iris')
#iris.head()

%matplotlib inline
sb.set()
sb.pairplot(df, hue='target_names', size=1.5)
```
